#CLUSTER ANALYSIS

```python
#Import the datasets
data_location = pd.read_csv("data_location.csv")
data_number = pd.read_csv("data_number.csv")
data_corner = pd.read_csv("data_corner.csv")
```

```python
#Loop through everything

final_coord = pd.DataFrame()

for i in range(len(data_number)):

  source_points = np.array([
        [data_corner["x"][i]+800,0],
        [data_corner["x"][i]+1400,0],
        [data_corner["x"][i]+1600, data_corner["y"][i]],
        [data_corner["x"][i], data_corner["y"][i]-50]
  ])
  destination_points = np.array([
        [0,0],
        [1300,0],
        [1300, 5000],
        [0,5000]
  ])

  h, status = cv2.findHomography(source_points, destination_points)

  #Create a vector of only 1s
  new_col = np.ones((data_location[data_location["ID"]==i+1].shape[0]))

  old_coordinates = np.matrix(data_location[data_location["ID"]==i+1].iloc[:,3:5])
  old_coordinates = np.column_stack((old_coordinates,new_col))


  adjusted_coordinates = np.transpose(h@np.transpose(old_coordinates))
  adjusted_coordinates[:,0] = np.divide(adjusted_coordinates[:,0],adjusted_coordinates[:,2])
  adjusted_coordinates[:,1] = np.divide(adjusted_coordinates[:,1],adjusted_coordinates[:,2])
  adjusted_coordinates = pd.DataFrame(adjusted_coordinates[:,0:2])

  final_coord = pd.concat([final_coord,adjusted_coordinates])
```

```python
data_location_adjusted = pd.DataFrame(np.column_stack((np.matrix(data_location.iloc[:,0:3]), np.matrix(final_coord))), columns=["ID", "Date", "Time", "x", "y"])
```

```python
#Check that it works
for i in range(2, 3):
  seaborn.set(style='ticks')
  plt.figure(figsize=(6,12))
  seaborn.scatterplot(x="x",y="y",data=data_location_adjusted.loc[data_location_adjusted['ID'] == i+1])
  plt.ylim(5000, 0)
  plt.xlim(0, 1300)
  plt.show()
```

```python
data_location_adjusted.to_csv("data_location_adjusted.csv",index=False)
```

#HOMOGRAPHY EXAMPLE

```python
im_src = cv2.imread('im_src.jpg')

data_corner_x = 38
data_corner_y = 782

source_points = np.array([
        [data_corner_x+800,0],
        [data_corner_x+1400,0],
        [data_corner_x+1600, data_corner_y],
        [data_corner_x, data_corner_y-50]
])

destination_points = np.array([
        [0,0],
        [1000,0],
        [1000, 2500],
        [0,2500]
])

h, status = cv2.findHomography(source_points, destination_points)


# Warp source image to destination based on homography
im_out = cv2.warpPerspective(im_src, h, (1000,2500))
```

```python
cv2_imshow(im_out)
```
